```python
### DECIDE ON HOW THIS DATA NEEDS TO BE SCALED
### Create Distrubution charts to decide

# Import library for Plotting distribution of numerical numbers
import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram for each numeric_column on df dataframe
def plot_distributions(df, numeric_columns):
    # Set up the plot size
    plt.figure(figsize=(15, 10))
    # Create a histogram for each numeric column start the enumeration at 1
    for c, column in enumerate(numeric_columns, 1):
        # Create the grid for each numeric_column 3x2 grid, 2 columns 3 rows
        plt.subplot(3, 2, c)
        # Use seaborn to create the histogram
        sns.histplot(data = df, x = column)
        # Title of each column is the column name itself
        plt.title(f'Plot {c}: Distribution of {column}')
    # Display layout spacing
    plt.tight_layout()
    # Display each histogram
    plt.show()

# Now actually call the function with your data:
plot_distributions(df, numeric_columns)
```
