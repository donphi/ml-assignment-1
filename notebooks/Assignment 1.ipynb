```python
# Make a copy of the dataframe for encoding
encoded_df = df.copy()

# Import Scikit Learn Label Encoder to convert categorical unique values to integers
from sklearn.preprocessing import LabelEncoder

# Run a for loop and apply Label Encoder to each categorical column
for column in categorical_columns:
    # First prints the name of the column in the loop
    print(f"Encoding Column: {column}")
    # Define the encoder for each loop for cleaning purposes
    encoder = LabelEncoder()
    # Second run the encoder through each column and label each unique column with unique integer
    encoded_df[column] = encoder.fit_transform(encoded_df[column])
    # Create a class_fit variable to add the classes to the encoder variable
    class_fit = encoder.classes_
    #
    class_transformed = encoder.transform(class_fit)
    for c in range(len(class_fit)):
        print(f"{class_fit[c]} -> {class_transformed[c]}")
    print("\n")
```
